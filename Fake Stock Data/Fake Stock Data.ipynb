```python
import random
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Simulating Fake Stock Returns

class Stock:
    
    def __init__(self,price,n):
        self.price = price
        self.priceMoves = self.priceMoves(n)
        self.series = self.getSeries()
        
    
    def priceMoves(self,n):  # Generates a list of priceMoves
        priceMoves = []
        for i in range(n):
            pos = 1 + (random.random()/100)
            neg = 1 - (random.random()/100)
            change = random.choice([pos,neg]) # Eqaul chance of down move or up move
            priceMoves.append(change)
        return priceMoves
        
    def getSeries(self):
        stockPrices = []
        initPrice = self.price
        for i in range(len(self.priceMoves)):
            newPrice = initPrice * self.priceMoves[i]
            stockPrices.append(newPrice)
            initPrice = newPrice
        return  stockPrices
```

```python
# Generating Fake Stock Data

returns = Stock(100,1000)
df = pd.DataFrame(
    {'prices' : returns.series
    }
)
```

```python
# Plotting the Fake Data

plt.plot(df.index, df['prices'])
plt.xlabel('Time Period')
plt.ylabel('Price')
plt.title('Plotting Fake Stock Data')
```
