```python
# Creating a list of Tickers
tickerList = list(major_indices['Symbol'].str.replace('^', '').str.lower())
```

```python
# Downloading data based on Tickers
df = yf.download(tickerList, period='1d', start='2020-01-13', end='2021-03-10' )
```

```python
# Cleaning up the data and adjusting for null values

adj_close = df.dropna(thresh=10, axis=1)['Adj Close']
```

```python
# Analysing summary statistics with the data
adj_close.describe().T
```

```python
# Subplots of Each Indice

adj_close.plot(figsize=(12,12), subplots=True);
```

```python
# Dropping the XAX index due to bad data
adj_close = adj_close.drop('XAX', axis=1)
```

```python
# Subplots after having dropped the index
adj_close.plot(figsize=(12,12), subplots=True)
```

```python
# Getthing the adjusted close prices of the dji only

dji = adj_close['DJI']
```

```python
# Getting Percentage Returns

dji_per_change = dji/ dji.shift(1) - 1
```

```python
dji_per_change.plot(figsize=(12,6))
```

```python
# Converting Percentage returns to log returns 

dji_log_returns_shift = np.log(dji/dji.shift(1))
```

```python
dji_log_returns_shift.plot(figsize=(12,6))
```

```python
# Histogram of closing prices

dji.hist(bins=50, figsize=(12,12))
```

```python
# Histogram of Percentage Returns

dji_per_change.hist(bins=50, figsize=(12,12))
```

```python
# Histogram of Log Returns 
dji_log_returns_shift.hist(bins=50, figsize=(12,12))
```
