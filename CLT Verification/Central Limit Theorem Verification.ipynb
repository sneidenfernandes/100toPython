```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generates a list of random data points that have been drawn from a uniform distribution to obtain a population
population_size = 100000 #Size of the population 
population = np.random.rand(population_size) 
```

```python
# Generates a distribution plot of the population

plt.hist(population,color='indianred',ec='black',bins=50)
plt.show()
```

```python
# Generates a random sample from the population 

sample_size = 1000
random_sample  = np.random.choice(population, size=sample_size)
```

```python
# Generates a distribution plot of the random sample

plt.hist(random_sample, color='pink',ec='black', bins=50)
plt.show()
```

```python
# Number of samples to generate
num_samples = 1000 
```

```python
# Generating a list of sample means

def sample_means(n):
    s_means = list()
    for i in range(num_samples):
        sample = np.random.choice(random_sample, size=n, replace=False)
        s_means.append(np.mean(sample))
    return s_means

# Generates a plot of sample means with n sample size
def plt_sample_means(n):
    sm = sample_means(n)
    pop_mean_estimator = np.mean(sm)
    plt.hist(sm, color='lightblue',ec='black',bins=50)
    plt.title(f" n = {n}")
    plt.figure(figsize=(1,1))
    plt.show() 

```

```python
# Generating Plots for increasing sample sizes

for i in range(1,100,10):
    plt_sample_means(i)
```
